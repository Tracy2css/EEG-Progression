```python
import pyxdf
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Load the XDF file
# data, header = pyxdf.load_xdf(r"F:\360MoveData\CurrentStudy\sub-P003\ses-S001\eeg\sub-P003_ses-S001_task-Default_run-001_eeg.xdf")
data, header = pyxdf.load_xdf(r"F:\360MoveData\CurrentStudy\sub-P002\ses-S001\eeg\sub-P002_ses-S001_task-Default_run-001_eeg.xdf")
```

```python
# Assuming the EEG data stream is the first stream
eeg_stream = data[1]
timestamps = eeg_stream['time_stamps']
samples = eeg_stream['time_series']
```

```python
# Create a DataFrame
df = pd.DataFrame(samples, columns=[f'Channel {i+1}' for i in range(samples.shape[1])])
df['Timestamps'] = timestamps
df.head(10)
```

```python
# Function to calculate the mean
def calculate_mean(series):
    return np.mean(series)

# Calculate max and mean for each column
max_values = df.max()
means = df.apply(calculate_mean)

# Create a summary DataFrame
summary_df = pd.DataFrame({'Max': max_values, 'Mean': means})

# Display the summary DataFrame
print(summary_df)
```

```python
# Save as a CSV file
df.to_csv('sub-P002_ses-S001_task-Default_run-001_eeg.csv', index=False)
```
